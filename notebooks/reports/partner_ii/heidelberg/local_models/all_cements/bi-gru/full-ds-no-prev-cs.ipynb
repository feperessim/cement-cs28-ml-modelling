```python
def flat_list(lst):
    return [item for tup in list(lst) for item in tup]
```

```python
def extract_numbers(text):
    values = []
    for line in text.split("\n"):
        line = line.strip()
        if line.startswith("RMSE:"):
            parts = line.split(" ")
            values.append(float(parts[1]))
            values.append(float(parts[2].lstrip("(").rstrip(")")))
        elif line.startswith("MAE:"):
            parts = line.split(" ")
            values.append(float(parts[1]))
            values.append(float(parts[2].lstrip("(").rstrip(")")))
        elif line.startswith("MAPE:"):
            parts = line.split(" ")
            values.append(float(parts[1]))
            values.append(float(parts[2].lstrip("(").rstrip(")")))
        elif line.startswith("R2:"):
            parts = line.split(" ")
            values.append(float(parts[1]))
            values.append(float(parts[2].lstrip("(").rstrip(")")))
    return values
```

```python
def extract_and_fill_lists(text, train_rmse_mean, train_rmse_std, test_rmse_mean, test_rmse_std,
                           train_mae_mean, train_mae_std, test_mae_mean, test_mae_std,
                           train_mape_mean, train_mape_std, test_mape_mean, test_mape_std,
                           train_r2_mean, train_r2_std, test_r2_mean, test_r2_std):

    train_flag = False
    test_flag = False
    timesteps_flag = False

    for line in text.split("\n"):
        line = line.strip()

        if line.startswith("TIMESTEPS:"):
            timesteps_flag = True

        elif line.startswith("[TRAIN]"):
            train_flag = True
            test_flag = False

        elif line.startswith("[TEST]"):
            train_flag = False
            test_flag = True

        elif line.startswith("RMSE:"):
            parts = line.split(" ")
            if train_flag:
                train_rmse_mean.append(float(parts[1]))
                train_rmse_std.append(float(parts[2].lstrip("(").rstrip(")")))
            elif test_flag:
                test_rmse_mean.append(float(parts[1]))
                test_rmse_std.append(float(parts[2].lstrip("(").rstrip(")")))

        elif line.startswith("MAE:"):
            parts = line.split(" ")
            if train_flag:
                train_mae_mean.append(float(parts[1]))
                train_mae_std.append(float(parts[2].lstrip("(").rstrip(")")))
            elif test_flag:
                test_mae_mean.append(float(parts[1]))
                test_mae_std.append(float(parts[2].lstrip("(").rstrip(")")))

        elif line.startswith("MAPE:"):
            parts = line.split(" ")
            if train_flag:
                train_mape_mean.append(float(parts[1]))
                train_mape_std.append(float(parts[2].lstrip("(").rstrip(")")))
            elif test_flag:
                test_mape_mean.append(float(parts[1]))
                test_mape_std.append(float(parts[2].lstrip("(").rstrip(")")))

        elif line.startswith("R2:"):
            parts = line.split(" ")
            if train_flag:
                train_r2_mean.append(float(parts[1]))
                train_r2_std.append(float(parts[2].lstrip("(").rstrip(")")))
            elif test_flag:
                test_r2_mean.append(float(parts[1]))
                test_r2_std.append(float(parts[2].lstrip("(").rstrip(")")))

        elif line.startswith("="):
            train_flag = False
            test_flag = False
            timesteps_flag = False

        else:
            continue
```
