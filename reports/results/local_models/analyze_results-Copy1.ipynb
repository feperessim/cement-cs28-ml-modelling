```python
def read_csv_files_grouped(csv_files, plant, header=[0, 1]):
    results = []
    for filepath in csv_files:
        df = pd.read_csv(filepath, header=header, low_memory=False)
        df["plant"] = plant  # Add the 'plant' column
        results.append(df)
    df = pd.concat(results, ignore_index=True)
    return df
```

```python
def read_csv_files_full(csv_files, plant):
    results = []
    for filepath in csv_files:
        df = pd.read_csv(filepath, low_memory=False)
        df["plant"] = plant  # Add the 'plant' column
        results.append(df)
    df = pd.concat(results, ignore_index=True)
    return df
```

### Renaming the Columns

```python
def preprocess_rename_columns(df):
    df.columns = pd.MultiIndex.from_tuples(
        list(
            {
                col: (col[0], "") if "Unnamed" in col[1] else col for col in df.columns
            }.values()
        )
    )
    return df
```

```python
def preprocess_change_columns_order(df, column, pos):
    # Get the list of columns
    cols = df.columns.tolist()
    # Remove the specified column
    cols.remove(column)
    # Insert the column at the desired position
    cols.insert(pos, column)
    # Reorder the DataFrame columns
    return df[cols]
```

```python
def read_csv_files_path(csv_files_path_dict, path, plant):
    csv_files = glob.glob(os.path.join(path, "*.csv"))
    csv_files_path_dict[plant] = csv_files
    return csv_files_path_dict
```

# Topological Example

```python
import pandas as pd
import numpy as np
from collections import deque, defaultdict

# Sample DataFrame containing the results for each model, including standard deviations
data = {
    'Model': ['Model1', 'Model2', 'Model3', 'Model4'],
    'RMSE_mean': [10, 15, 12, 8],
    'RMSE_std': [1, 2, 1.5, 0.8],
    'MAE_mean': [5, 7, 6, 4],
    'MAE_std': [0.5, 0.7, 0.6, 0.4],
    'MAPE_mean': [3, 4, 3.5, 2.5],
    'MAPE_std': [0.3, 0.4, 0.35, 0.25],
    'R2_mean': [0.9, 0.85, 0.88, 0.92],
    'R2_std': [0.05, 0.07, 0.06, 0.04]
}

df = pd.DataFrame(data)

# Function to compare two models strictly
def strictly_dominates(model1, model2):
    """Return True if model1 strictly dominates model2, False otherwise."""
    return (model1['RMSE_mean'] < model2['RMSE_mean'] and model1['RMSE_std'] < model2['RMSE_std'] and
            model1['MAE_mean'] < model2['MAE_mean'] and model1['MAE_std'] < model2['MAE_std'] and
            model1['MAPE_mean'] < model2['MAPE_mean'] and model1['MAPE_std'] < model2['MAPE_std'] and
            model1['R2_mean'] > model2['R2_mean'] and model1['R2_std'] < model2['R2_std'])

# Create the dominance matrix
n = len(df)
dominance_matrix = np.zeros((n, n), dtype=bool)

for i in range(n):
    for j in range(n):
        if i != j:
            dominance_matrix[i, j] = strictly_dominates(df.iloc[i], df.iloc[j])

# Create the dominance graph
dominance_graph = {i: [] for i in range(n)}
for i in range(n):
    for j in range(n):
        if dominance_matrix[i, j]:
            dominance_graph[i].append(j)

# Topological Sorting
def topological_sort(graph):
    """Perform topological sorting on the given graph."""
    in_degree = {u: 0 for u in graph}
    for u in graph:
        for v in graph[u]:
            in_degree[v] += 1
    
    queue = deque([u for u in graph if in_degree[u] == 0])
    topo_order = []
    
    while queue:
        u = queue.popleft()
        topo_order.append(u)
        
        for v in graph[u]:
            in_degree[v] -= 1
            if in_degree[v] == 0:
                queue.append(v)
    
    return topo_order

# Get the topological order
topo_order = topological_sort(dominance_graph)

# Categorize models based on dominance
dominates_any = np.any(dominance_matrix, axis=1)
dominated_by_any = np.any(dominance_matrix, axis=0)

dominating_models = [df['Model'].iloc[i] for i in range(n) if dominates_any[i] and not dominated_by_any[i]]
non_dominated_models = [df['Model'].iloc[i] for i in range(n) if not dominates_any[i] and not dominated_by_any[i]]
failing_models = [df['Model'].iloc[i] for i in range(n) if dominated_by_any[i]]

print("Dominating Models:", dominating_models)
print("Non-Dominated Models:", non_dominated_models)
print("Failing Models:", failing_models)

# Map the topological order to the model names
sorted_models = [df['Model'].iloc[i] for i in topo_order]
print("Models in order of dominance:", sorted_models)

# Visualization of the dominance graph
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph(dominance_graph)
labels = {i: df['Model'].iloc[i] for i in range(n)}
pos = nx.spring_layout(G)
nx.draw(G, pos, labels=labels, with_labels=True, node_size=3000, node_color="skyblue", font_size=12, font_color="black", font_weight="bold", arrows=True)
plt.title("Dominance Graph")
plt.show()
```

```python
# Count how many models each model dominates and is dominated by
dominates_count = np.sum(dominance_matrix, axis=1)
dominated_count = np.sum(dominance_matrix, axis=0)

# Add these counts to the DataFrame
df['Dominates_Count'] = dominates_count
df['Dominated_Count'] = dominated_count

# Models that dominate others but are not dominated
df['Dominates_Others'] = df['Dominates_Count'] > 0
df['Not_Dominated'] = df['Dominated_Count'] == 0

# Summary statistics
dominance_summary = {
    'Total Models': len(df),
    'Dominating Models': df['Dominates_Others'].sum(),
    'Non-Dominated Models': df['Not_Dominated'].sum(),
    'Failing Models': df['Dominated_Count'].sum()
}

print(dominance_summary)
```

```python
import pandas as pd
import numpy as np
from collections import deque, defaultdict


# Function to compare two models strictly
def strictly_dominates(model1, model2):
    """Return True if model1 strictly dominates model2, False otherwise."""
    return (
        model1["RMSE_mean"] < model2["RMSE_mean"]
        and model1["RMSE_std"] < model2["RMSE_std"]
        and model1["MAE_mean"] < model2["MAE_mean"]
        and model1["MAE_std"] < model2["MAE_std"]
        and model1["MAPE_mean"] < model2["MAPE_mean"]
        and model1["MAPE_std"] < model2["MAPE_std"]
        and model1["R2_mean"] > model2["R2_mean"]
        and model1["R2_std"] < model2["R2_std"]
    )

    # Function to determine strict dominance


def strictly_dominates(arr1, arr2):
    return (
        (arr1[:, None, 0] < arr2[:, 0])
        & (arr1[:, None, 1] < arr2[:, 1])  # RMSE_mean
        & (arr1[:, None, 2] < arr2[:, 2])  # RMSE_std
        & (arr1[:, None, 3] < arr2[:, 3])  # MAE_mean
        & (arr1[:, None, 4] < arr2[:, 4])  # MAE_std
        & (arr1[:, None, 5] < arr2[:, 5])  # MAPE_mean
        & (arr1[:, None, 6] > arr2[:, 6])  # MAPE_std
        & (arr1[:, None, 7] < arr2[:, 7])  # R2_mean  # R2_std
    )


# def strictly_dominates(arr1, arr2):
#     return (
#         (arr1[:, None, 0] < arr2[:, 0]) &  # RMSE_mean
#         (arr1[:, None, 1] < arr2[:, 1]) &  # RMSE_cv
#         (arr1[:, None, 2] < arr2[:, 2]) &  # MAE_mean
#         (arr1[:, None, 3] < arr2[:, 3]) &  # MAE_cv
#         (arr1[:, None, 4] < arr2[:, 4]) &  # MAPE_mean
#         (arr1[:, None, 5] < arr2[:, 5]) &  # MAPE_cv
#         (arr1[:, None, 6] > arr2[:, 6]) &  # R2_mean
#         (arr1[:, None, 7] < arr2[:, 7])    # R2_cv
#     )


# Function to find the top models in each group
def find_top_models(group):
    max_net_dominance = group["Net_Dominance"].max()
    top_models = group[group["Net_Dominance"] == max_net_dominance]
    return top_models


# Topological Sorting
def topological_sort(graph):
    """Perform topological sorting on the given graph."""
    in_degree = {u: 0 for u in graph}
    for u in graph:
        for v in graph[u]:
            in_degree[v] += 1

    queue = deque([u for u in graph if in_degree[u] == 0])
    topo_order = []

    while queue:
        u = queue.popleft()
        topo_order.append(u)

        for v in graph[u]:
            in_degree[v] -= 1
            if in_degree[v] == 0:
                queue.append(v)

    return topo_order
```

```python
## df_results df_results.T.rename()

df_results = df_copy[
    [
        ("Company", ""),
        ("plant", ""),
        ("Cross Validation", ""),
        ("Features", ""),
        ("Model", ""),
        ("Model_bkp", ""),
        ("Timesteps", ""),
        ("RMSE Test", "mean"),
        ("MAE Test", "mean"),
        ("MAPE Test", "mean"),
        ("R2 Test", "mean"),
        ("RMSE Test", "std"),
        ("MAE Test", "std"),
        ("MAPE Test", "std"),
        ("R2 Test", "std"),
    ]
]
df_results = df_results.reset_index(drop=True)

# Rename columns
new_column_names = [
    "Company",
    "Plant",
    "Cross Validation",
    "Features",
    "Model",
    "Model_bkp",
    "Timesteps",
    "RMSE_mean",
    "MAE_mean",
    "MAPE_mean",
    "R2_mean",
    "RMSE_std",
    "MAE_std",
    "MAPE_std",
    "R2_std",
]

df_results.columns = new_column_names

df_results["Model_bkp2"] = df_results["Model_bkp"] + df_results["Timesteps"].fillna(
    -1
).astype(int).apply(lambda x: "_" + str(x) if x != -1 else "").astype(str)

# Normalize standard deviations using the coefficient of variation (CV)
df_results["RMSE_cv"] = df_results["RMSE_std"] / df_results["RMSE_mean"]
df_results["MAE_cv"] = df_results["MAE_std"] / df_results["MAE_mean"]
df_results["MAPE_cv"] = df_results["MAPE_std"] / df_results["MAPE_mean"]
df_results["R2_cv"] = df_results["R2_std"] / df_results["R2_mean"]
```

```python
# Initialize lists to store the dominance counts
dominates_list = []
dominated_by_list = []
dominant_model_list = []

# Group by Cross Validation, Features, and Plant
grouped = df_results.groupby(["Cross Validation", "Features", "Plant"])

# Loop through each group
for name, group in grouped:
    n = len(group)
    group_values = group[
        [
            "RMSE_mean",
            "RMSE_std",
            "MAE_mean",
            "MAE_std",
            "MAPE_mean",
            "MAPE_std",
            "R2_mean",
            "R2_std",
        ]
    ].values

    #     group_values = group[
    #         [
    #             "RMSE_mean",
    #             "RMSE_cv",
    #             "MAE_mean",
    #             "MAE_cv",
    #             "MAPE_mean",
    #             "MAPE_cv",
    #             "R2_mean",
    #             "R2_cv",
    #         ]
    #     ].values

    # Calculate the dominance matrix using vectorized operations
    dominance_matrix = strictly_dominates(group_values, group_values)

    # Exclude self-comparisons
    # Necessary only for not stricktly comparisons.
    # np.fill_diagonal(dominance_matrix, False)

    # Calculate dominance counts for each model in the group
    dominates_count = dominance_matrix.sum(axis=1)
    dominated_by_count = dominance_matrix.sum(axis=0)

    # Identify dominant models
    dominant_models = np.where(dominated_by_count == 0)[0]

    # Store the results with the group
    for i in range(n):
        dominates_list.append(dominates_count[i])
        dominated_by_list.append(dominated_by_count[i])
        dominant_model_list.append(i in dominant_models)
```

```python
df_results["Dominates"] = pd.Series(dominates_list, index=df_results.index)
df_results["Dominated_By"] = pd.Series(dominated_by_list, index=df_results.index)
df_results["Is_Dominant"] = pd.Series(dominant_model_list, index=df_results.index)
df_results["Net_Dominance"] = df_results["Dominates"] - df_results["Dominated_By"]
```
